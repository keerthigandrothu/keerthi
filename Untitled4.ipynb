```python
import pandas as pd
import matplotlib.pyplot as plt

# Create advanced dataset
data = {
    "Product": ["A","B","C","A","B","C","A","B","C"],
    "Month": ["Jan","Jan","Jan","Feb","Feb","Feb","Mar","Mar","Mar"],
    "Sales": [100,150,200,120,180,220,130,170,250]
}

df = pd.DataFrame(data)

print("Dataset:\n", df)

# Total Sales
total_sales = df["Sales"].sum()
print("\nTotal Sales:", total_sales)

# Product-wise Sales
product_sales = df.groupby("Product")["Sales"].sum()
print("\nSales by Product:\n", product_sales)

# Month-wise Sales
monthly_sales = df.groupby("Month")["Sales"].sum()
print("\nSales by Month:\n", monthly_sales)

# Best Selling Product
best_product = product_sales.idxmax()
print("\nBest Selling Product:", best_product)

# 📊 Graph 1: Product Sales
product_sales.plot(kind='bar')
plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()

# 📈 Graph 2: Monthly Trend
monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
