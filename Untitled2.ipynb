```python
print("----- Student Report -----")

name = "keerthi"
sub1 = 90
sub2 = 80
sub3 = 70

total = sub1 + sub2 + sub3
average = total / 3

if average >= 90:
    grade = "A"
elif average >= 75:
    grade = "B"
elif average >= 50:
    grade = "C"
else:
    grade = "Fail"

print("\n----- Result -----")
print("Name:", name)
print("Total:", total)
print("Average:", average)
print("Grade:", grade)
```
